```python
import pandas as pd
import numpy as np
import plotly
import plotly.graph_objects as go
data = pd.read_csv('Data.csv')

data_broadband = data.dropna(subset = ['Country','Broadband Speed Rank','Broadband Mbps'])
data_mobile = data.dropna(subset = ['Country','Mobile Speed Rank','Mobile Mbps'])

data_broadband.to_csv('Broadband.csv')
data_mobile.to_csv('Mobile.csv')


import plotly.express as px

fig1  = px.choropleth(data_broadband, 
            locations='Country',
             color="Broadband Mbps",
            color_continuous_scale="Inferno",
              locationmode="country names",
               #scope="usa",
             range_color=(0,200),
             height=500)

fig1.update_layout(
    title="Global Broadband Speed Rank",
    legend_title="Broadband Speed (Mbps)",
    font=dict(
        #family="Courier New, monospace",
        #size=18,
        #color="RebeccaPurple"
    )
)

fig1.show()
fig1.write_image('Broadband.jpeg')

fig2  = px.choropleth(data_mobile, 
            locations='Country',
             color="Mobile Mbps",
            color_continuous_scale="Inferno",
              locationmode="country names",
               #scope="usa",
             range_color=(0,140),
             title='Global Mobile Speed Rank',
             height=500)

fig2.show()
fig2.write_image('Mobile_Speed.jpeg')
```

```python
import matplotlib.pyplot as plt
max_broadband = data_broadband.nlargest(n=10,columns = ['Broadband Mbps'])
plt.figure(figsize = (10, 5))
plt.rcParams['font.size'] = '16'
# plt.bar(max_broadband['Country'],max_broadband['Broadband Mbps'], width = .4)
y = max_broadband['Country']
x = max_broadband['Broadband Mbps']
plt.barh(y,x)
plt.xlabel("Broadband Mbps, Top Ten", fontsize = 20)
plt.ylabel("Country", fontsize = 20)
# plt.savefig('Top_Ten.jpeg', bbox_inches="tight")
```

```python
max_broadband = data_broadband.nsmallest(n=10,columns = ['Broadband Mbps'])
plt.figure(figsize = (10, 5))
plt.rcParams['font.size'] = '16'
# plt.bar(max_broadband['Country'],max_broadband['Broadband Mbps'], width = .4)
x = max_broadband['Country']
y = max_broadband['Broadband Mbps']
plt.barh(x,y)
plt.xlabel("Broadband Mbps, Lowest Ten", fontsize = 20)
plt.ylabel("Country" , fontsize = 20)
plt.savefig('Lowest_Ten.jpeg', bbox_inches="tight")
```
